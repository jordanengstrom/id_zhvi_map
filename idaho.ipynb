```python
import pandas as pd
import json
```

```python
id_df = pd.read_csv('./zhvi_data/zip_zhvi_2022.csv')
# state_zhvi_df.shape
# (30523, 275)
id_df = id_df.query('StateName == "ID"')
id_df.reset_index(drop=True, inplace=True)
# id_df.shape
# (262, 275)

zip_df = pd.read_csv('./zhvi_data/zipcodes.csv')
id_zip_codes = id_df['RegionName'].unique()
print(f'total number of idaho zip codes: {len(id_zip_codes)}\n')

id_df = id_df.merge(right=zip_df, how='inner', left_on='RegionName', right_on='ZIP')
id_df.head()
# Column `RegionName` is zip code
# print(f'we have {len(id_df)} rows:')
# str(id_df.columns.tolist())
```
